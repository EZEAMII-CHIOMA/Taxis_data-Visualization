```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
taxis_data = sns.load_dataset("taxis")
```

```python
taxis_data
```

```python
# Univariate Visualizations
# 1. Histogram of the 'fare' column
plt.figure(figsize=(10, 5))
sns.histplot(taxis_data['fare'], bins=20, kde=True)
plt.title("Distribution of Fare Amount")
plt.xlabel("Fare Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
2. #Countplot of the 'payment' method
plt.figure(figsize=(8, 5))
sns.countplot(data=taxis_data, x='payment')
plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bivariate Visualizations
# 1. Scatter plot of 'distance' vs. 'fare'
plt.figure(figsize=(8, 5))
sns.scatterplot(data=taxis_data, x='distance', y='fare', hue='payment')
plt.title("Distance vs. Fare")
plt.xlabel("Distance")
plt.ylabel("Fare")
plt.legend(title='Payment')
plt.show()
```

```python
# 2. Boxplot of 'passengers' and 'fare'
plt.figure(figsize=(8, 5))
sns.boxplot(data=taxis_data, x='passengers', y='fare')
plt.title("Passengers vs. Fare")
plt.xlabel("Passengers")
plt.ylabel("Fare")
plt.show()
```

```python
# Multivariate Visualizations
# plt.figure(figsize=(8, 6))
correlation_matrix = taxis_data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 2. Violin plot of 'pickup_borough' vs. 'fare'
plt.figure(figsize=(12, 6))
sns.violinplot(data=taxis_data, x='pickup_borough', y='fare', palette='Set2')
plt.title("Fare Distribution by Pickup Borough")
plt.xlabel("Pickup Borough")
plt.ylabel("Fare")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3.
g = sns.pairplot(
    taxis_data,
    vars=['distance', 'fare', 'tip'],
    hue='payment',
    markers=['o', 's'],
    plot_kws={'alpha': 0.7}
)
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Pairplot of Distance, Fare, and Tip by Payment Method")
plt.show()
```
